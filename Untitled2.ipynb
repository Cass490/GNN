```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import networkx as nx
import json
import numpy as np
import os
import random

# Determine device (GPU if available, otherwise CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")


class GCNLayer(nn.Module):
    def __init__(self, in_features, out_features):
        super(GCNLayer, self).__init__()
        self.linear = nn.Linear(in_features, out_features)

    def forward(self, node_features, adj_matrix):
        # Normalize adjacency matrix
        deg = torch.sum(adj_matrix, dim=1)
        deg_inv_sqrt = torch.pow(deg + 1e-6, -0.5)
        norm_adj = torch.mul(
            torch.mul(adj_matrix, deg_inv_sqrt.unsqueeze(1)),
            deg_inv_sqrt.unsqueeze(0)
        )
        
        # Perform graph convolution
        aggregated_features = torch.matmul(norm_adj, node_features)
        transformed_features = self.linear(aggregated_features)
        return transformed_features


class MyGCN(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(MyGCN, self).__init__()
        self.gcn1 = GCNLayer(input_dim, hidden_dim)
        self.gcn2 = GCNLayer(hidden_dim, output_dim)
        self.dropout = nn.Dropout(0.2)
        
    def forward(self, adj_matrix, node_features):
        # First GCN layer with ReLU and dropout
        h1 = F.relu(self.gcn1(node_features, adj_matrix))
        h1 = self.dropout(h1)
        # Second GCN layer
        h2 = self.gcn2(h1, adj_matrix)
        return h2

class MyHeuristic():
    def __init__(self, params):
        self.params = params

    def __call__(self, gnn_embeddings, raw_features, available_jobs, graph = None):
        #Implement your heuristic function
        priorities = {}
        for job_id in available_jobs:
            priority = raw_features["jobs"][f"job_{job_id}"]["priority"]
            priorities[f"job_{job_id}"]=priority
        if priorities:
            action = max(priorities, key = priorities.get)
        else:
            action = None
        return action

class FMSDatasetGenerator:
    """Generates datasets for an FMS scheduling problem."""

    def __init__(self, num_jobs, num_machines, num_ops_per_job_range, processing_time_range,
                 failure_rate, setup_time_range, resource_capacity, load_factor,
                 dynamic_arrival=True, fixed_routing=False):
        """
        Initializes the generator.
        Args:
            num_jobs (int or tuple): The number of jobs, or a range for the number of jobs
            num_machines (int): The number of machines.
            num_ops_per_job_range (tuple): Range of operations a job can have.
            processing_time_range (tuple): Processing time range (min, max).
            failure_rate (float): Probability of machine failure (per simulation step).
            setup_time_range (tuple): range for setup times (min, max).
            resource_capacity(int): number of resources shared by machines
            load_factor (float): Load factor for resource contention(percentage of machines that need shared resources).
            dynamic_arrival (bool): Whether to add dynamic arrivals
            fixed_routing (bool): whether the routing for each job is fixed
        """
        self.num_jobs = num_jobs if isinstance(num_jobs, int) else random.randint(*num_jobs)
        self.num_machines = num_machines
        self.num_ops_per_job_range = num_ops_per_job_range
        self.processing_time_range = processing_time_range
        self.failure_rate = failure_rate
        self.setup_time_range = setup_time_range
        self.resource_capacity = resource_capacity
        self.load_factor = load_factor
        self.dynamic_arrival = dynamic_arrival
        self.fixed_routing = fixed_routing
        self.machines = self._generate_machines()
        self.jobs = self._generate_jobs()
        self.resources = self._generate_resources()

    def _generate_resources(self):
        resources = {
            "resource_capacity": self.resource_capacity,
            "machines":  random.sample(self.machines, int(len(self.machines) * self.load_factor))
        }
        return resources

    def _generate_machines(self):
        """Generates machine data."""
        machines = []
        for machine_id in range(self.num_machines):
            machines.append({"machine_id": machine_id,
                               "status": "available",  # initially available
                               "current_job": None,
                               "next_available_time": 0,
                               "type": random.choice(["general", "special"])  # Example of different machine types
                              })
        return machines

    def _generate_jobs(self):
        """Generates job data."""
        jobs = []
        for job_id in range(self.num_jobs):
            num_ops = random.randint(*self.num_ops_per_job_range)
            ops = self._generate_job_ops(job_id, num_ops)
            jobs.append({"job_id": job_id, "operations": ops,
                           "arrival_time": 0 if not self.dynamic_arrival else random.randint(0,50),
                           "completed_operations":0,
                           "priority": random.randint(1,5),
                           "machine_seq": [op["machine"] for op in ops],
                           "current_operation":0
                           })
        return jobs

    def _generate_job_ops(self, job_id, num_ops):
        """Generates operations for a single job."""
        ops = []
        available_machines = self._get_available_machines()
        if self.fixed_routing:
            machine_seq = random.sample(available_machines, num_ops)
        else:
            machine_seq = []

        for op_id in range(num_ops):
            if self.fixed_routing:
                machine = machine_seq[op_id]
            else:
                machine = random.choice(available_machines)
            proc_time = self._triangular_distribution(*self.processing_time_range)
            setup_time = random.uniform(*self.setup_time_range) if op_id > 0 else 0.0
            ops.append({"job_id": job_id, "op_id": op_id, "machine": machine,
                            "processing_time": proc_time, "setup_time": setup_time,
                            "status": "pending"  # initially pending
                            })
        return ops

    def _triangular_distribution(self, min, max):
        mode = (min + max)/2
        u = random.uniform(0,1)
        if u <= (mode-min)/(max-min):
            return min + np.sqrt(u*(max-min)*(mode-min))
        else:
            return max-np.sqrt((1-u)*(max-min)*(max-mode))

    def _get_available_machines(self):
        available_machines = []
        for m in self.machines:
            available_machines.append(m["machine_id"])
        return available_machines


    def save_dataset(self, filename, save_dir):
        """Saves dataset to a JSON file."""
        os.makedirs(save_dir, exist_ok = True)
        dataset = {
            "jobs": self.jobs,
            "machines": self.machines,
            "resources": self.resources,
        }

        with open(os.path.join(save_dir,filename), 'w') as f:
            json.dump(dataset, f, indent=4)

    @classmethod
    def load_dataset(cls, filename):
        """Loads dataset from a JSON file."""
        with open(filename, 'r') as f:
            dataset = json.load(f)
        return dataset

def create_graph_and_features(dataset, current_time = 0):
    """
        Creates graph data and feature vectors from current dataset status.
        Args:
            dataset (dict): The loaded dataset from JSON
            current_time (int): current time of the simulation
        Returns:
            tuple: (graph_data (nx.Graph), heuristic_data (dict))
        """
    jobs = dataset.jobs
    machines = dataset.machines
    graph = nx.DiGraph()
    heuristic_data = {"jobs": {}, "machines": {}, "operations": {}}
    # 1. Node generation:
    # Add job nodes
    for job in jobs:
        graph.add_node(f"job_{job['job_id']}", type="job", **job)
        heuristic_data["jobs"][f"job_{job['job_id']}"] = _extract_job_features(job,current_time)

    # Add machine nodes
    for machine in machines:
    # Create a copy without 'type' or any other non-node property
        machine_node_data = {k: v for k, v in machine.items() if k not in ["type", "current_job"]}
        graph.add_node(f"machine_{machine['machine_id']}", type="machine", **machine_node_data)
        heuristic_data["machines"][f"machine_{machine['machine_id']}"] = _extract_machine_features(machine, current_time)

    # Add operation nodes
    for job in jobs:
        for op in job["operations"]:
            graph.add_node(f"op_{job['job_id']}_{op['op_id']}", type="operation", **op)
            heuristic_data["operations"][f"op_{job['job_id']}_{op['op_id']}"] = _extract_op_features(op, job, current_time)

    # 2. Edge Generation
    #Edges for operations in a job sequence
    for job in jobs:
        ops = job["operations"]
        for i in range(len(ops)-1):
            graph.add_edge(f"op_{job['job_id']}_{ops[i]['op_id']}",
                        f"op_{job['job_id']}_{ops[i+1]['op_id']}", type="op_seq")

    #Edges between operations and machines
    for job in jobs:
        for op in job["operations"]:
            machine_node = f"machine_{op['machine']}"
            graph.add_edge(f"op_{job['job_id']}_{op['op_id']}", machine_node, type="machine_link")

    #Disjunctive edges for operations in a machine
    for m in machines:
        ops_in_machine = []
        for job in jobs:
            for op in job["operations"]:
                if op["machine"] == m["machine_id"]:
                    ops_in_machine.append(f"op_{job['job_id']}_{op['op_id']}")
        for i in range(len(ops_in_machine)):
            for j in range(i+1, len(ops_in_machine)):
                graph.add_edge(ops_in_machine[i], ops_in_machine[j], type = "disjunctive_link")

    return graph, heuristic_data

def _extract_job_features(job, current_time):
    """
    Extracts a feature vector for job.
    Args:
        job (dict): current job
        current_time (int): current simulation time
    Returns:
        dict: features
    """
    time_since_arrival = current_time-job["arrival_time"]
    remaining_operations = len(job["operations"])- job["completed_operations"]
    features = {"job_id": float(job["job_id"]),
                "priority": float(job["priority"]),
                "time_since_arrival": float(time_since_arrival),
                "remaining_operations":float(remaining_operations),
                "completed_operations": float(job["completed_operations"])
                }
    return features

def _extract_machine_features(machine, current_time):
    """Extracts a feature vector for machine"""
    features = { "machine_id": float(machine["machine_id"]),
                "next_available_time": float(machine["next_available_time"] - current_time )
            }
    return features

def _extract_op_features(op, job, current_time):
        """Extracts a feature vector for operation."""
        features = { "job_id": float(job["job_id"]),
                    "op_id": float(op["op_id"]),
                    "machine": float(op["machine"]),
                    "processing_time": float(op["processing_time"]),
                    "setup_time": float(op["setup_time"]),

        }
        return features

def prepare_gnn_input(graph, feature_data):
    """Prepares data for GNN with consistent feature dimensions."""
    # Define feature sizes for each node type
    job_feature_size = 5  # job_id, priority, time_since_arrival, remaining_ops, completed_ops
    machine_feature_size = 2  # machine_id, next_available_time
    operation_feature_size = 5  # job_id, op_id, machine, processing_time, setup_time
    
    # Use the maximum feature size for all nodes to ensure consistent dimensions
    max_feature_size = max(job_feature_size, machine_feature_size, operation_feature_size)
    
    node_features = []
    node_ids = []
    
    # Process all nodes and pad features to max_feature_size
    for node_id, node_data in graph.nodes(data=True):
        node_ids.append(node_id)
        
        if node_data["type"] == "job":
            features = list(feature_data["jobs"][f"job_{node_data['job_id']}"].values())
        elif node_data["type"] == "machine":
            features = list(feature_data["machines"][f"machine_{node_data['machine_id']}"].values())
        elif node_data["type"] == "operation":
            features = list(feature_data["operations"][f"op_{node_data['job_id']}_{node_data['op_id']}"].values())
        
        # Pad with zeros to reach max_feature_size
        padded_features = features + [0.0] * (max_feature_size - len(features))
        node_features.append(padded_features)
    
    # Convert to tensors
    node_features_tensor = torch.tensor(node_features, dtype=torch.float32)
    adj_matrix = nx.to_numpy_array(graph, nodelist=node_ids)
    adj_matrix_tensor = torch.tensor(adj_matrix, dtype=torch.float32)
    
    return node_features_tensor, adj_matrix_tensor

def run_scheduling_step(dataset, current_time, gnn_model, heuristic_algo):
    graph, feature_data = create_graph_and_features(dataset, current_time)
    available_jobs = [job["job_id"] for job in dataset.jobs 
                     if job["current_operation"] < len(job["operations"])]
    
    node_features, adj_matrix = prepare_gnn_input(graph, feature_data)
    
    # Move tensors to the appropriate device
    node_features = node_features.to(device)
    adj_matrix = adj_matrix.to(device)
    
    gnn_embeddings = gnn_model(adj_matrix, node_features)
    
    # Pass the graph to the heuristic
    action = heuristic_algo(gnn_embeddings, feature_data, available_jobs, graph)
    
    # Clear tensors that are not needed anymore
    del node_features
    del adj_matrix
    
    return action, dataset

class CentralityHeuristic():
    def __init__(self, params=None):
        self.params = params
        self.waiting_time_weight = 0.2
        self.priority_weight = 0.2
        self.centrality_weight = 0.4  # High weight for centrality
        self.processing_time_weight = 0.2
        
    def calculate_centrality_metrics(self, graph):
        """Calculate various centrality measures for the graph."""
        centrality_scores = {}
        
        # Degree centrality - how many direct connections?
        degree_cent = nx.degree_centrality(graph)
        
        # Betweenness centrality - how often node lies on shortest path?
        betweenness_cent = nx.betweenness_centrality(graph)
        
        # Eigenvector centrality - how important are your neighbors?
        try:
            eigenvector_cent = nx.eigenvector_centrality(graph)
        except:
            eigenvector_cent = {node: 0 for node in graph.nodes()}  # Fallback
            
        # Combine centrality measures
        for node in graph.nodes():
            centrality_scores[node] = {
                'degree': degree_cent[node],
                'betweenness': betweenness_cent[node],
                'eigenvector': eigenvector_cent[node],
                'combined': (degree_cent[node] + 
                           betweenness_cent[node] + 
                           eigenvector_cent[node]) / 3.0
            }
            
        return centrality_scores

    def __call__(self, gnn_embeddings, raw_features, available_jobs, graph):
        if not available_jobs:
            return None
            
        # Calculate centrality metrics
        centrality_scores = self.calculate_centrality_metrics(graph)
        
        job_scores = {}
        for job_id in available_jobs:
            job_key = f"job_{job_id}"
            job_data = raw_features["jobs"][job_key]
            
            # 1. Priority Score (1-5 from original data)
            priority_score = job_data["priority"] / 5.0
            
            # 2. Waiting Time Score
            waiting_time = job_data["time_since_arrival"]
            waiting_score = min(waiting_time / 50.0, 1.0)
            
            # 3. Centrality Score
            job_centrality = centrality_scores[job_key]['combined']
            centrality_score = job_centrality  # Already normalized
            
            # 4. Processing Time Score
            current_op_id = int(job_data["completed_operations"])
            op_key = f"op_{job_id}_{current_op_id}"
            if op_key in raw_features["operations"]:
                op_data = raw_features["operations"][op_key]
                proc_time = op_data["processing_time"]
                processing_score = 1.0 - (proc_time / 20.0)
            else:
                processing_score = 0
            
            # Calculate weighted total score
            total_score = (
                self.priority_weight * priority_score +
                self.waiting_time_weight * waiting_score +
                self.centrality_weight * centrality_score +
                self.processing_time_weight * processing_score
            )
            
            job_scores[job_key] = total_score
            
            # Print decision factors
            print(f"\nScoring {job_key}:")
            print(f"  Priority Score: {priority_score:.2f}")
            print(f"  Waiting Score: {waiting_score:.2f}")
            print(f"  Centrality Score: {centrality_score:.2f}")
            print(f"  Processing Score: {processing_score:.2f}")
            print(f"  Total Score: {total_score:.2f}")
            print(f"  Centrality Details:")
            print(f"    Degree: {centrality_scores[job_key]['degree']:.2f}")
            print(f"    Betweenness: {centrality_scores[job_key]['betweenness']:.2f}")
            print(f"    Eigenvector: {centrality_scores[job_key]['eigenvector']:.2f}")
        
        if job_scores:
            selected_job = max(job_scores, key=job_scores.get)
            print(f"\nSelected {selected_job} with highest score!")
            return selected_job
        return None

def update_system_state(dataset, selected_job, current_time):
    """Updates the system state with real job processing."""
    if selected_job is None:
        # Just update machine states if no job selected
        for machine in dataset.machines:
            if machine['next_available_time'] <= current_time:
                machine['status'] = 'available'
                machine['current_job'] = None
        return dataset

    job_id = int(selected_job.split('_')[1])
    
    # Find the selected job
    for job in dataset.jobs:
        if job['job_id'] == job_id:
            # Check if job still has operations
            if job['current_operation'] >= len(job['operations']):
                return dataset
                
            current_op = job['operations'][job['current_operation']]
            machine_id = current_op['machine']
            
            # Find the required machine
            for machine in dataset.machines:
                if machine['machine_id'] == machine_id and machine['status'] == 'available':
                    # Start processing the job
                    processing_time = current_op['processing_time']
                    setup_time = current_op['setup_time']
                    total_time = processing_time + setup_time
                    
                    # Update machine status
                    machine['status'] = 'busy'
                    machine['current_job'] = job_id
                    machine['next_available_time'] = current_time + total_time
                    
                    # Update operation status
                    current_op['status'] = 'completed'
                    
                    # Move to next operation
                    job['current_operation'] += 1
                    job['completed_operations'] += 1
                    
                    print(f"\nTime {current_time}:")
                    print(f"- Started job {job_id} operation {current_op['op_id']} on machine {machine_id}")
                    print(f"- Processing time: {processing_time}, Setup time: {setup_time}")
                    print(f"- Machine will be available at time {machine['next_available_time']}")
                    break
    
    # Update all machine states
    for machine in dataset.machines:
        if machine['next_available_time'] <= current_time:
            machine['status'] = 'available'
            machine['current_job'] = None
            
    return dataset

def main_simulation(train_data_dir, val_data_dir, test_data_dir, gnn_model, heuristic_algo, train_iters):
    """Main simulation loop with detailed progress tracking."""
    for i in range(train_iters):
        for filename in os.listdir(train_data_dir):
            print(f"\nProcessing dataset: {filename}")
            
            # Load dataset
            loaded_generator = FMSDatasetGenerator.load_dataset(os.path.join(train_data_dir, filename))
            
            # Initialize simulation
            current_time = 0
            num_steps = 20
            current_dataset = FMSDatasetGenerator(
                num_jobs=len(loaded_generator["jobs"]),
                num_machines=len(loaded_generator["machines"]),
                num_ops_per_job_range=(3, 8),
                processing_time_range=(2, 20),
                failure_rate=0.05,
                setup_time_range=(0, 3),
                resource_capacity=6,
                load_factor=0.6
            )
            current_dataset.jobs = loaded_generator["jobs"]
            current_dataset.machines = loaded_generator["machines"]
            current_dataset.resources = loaded_generator["resources"]
            
            # Print initial state
            print("\nInitial state:")
            print(f"Total jobs: {len(current_dataset.jobs)}")
            print(f"Total machines: {len(current_dataset.machines)}")
            
            # Simulation loop
            while current_time < num_steps:
                # Print periodic status updates
                if current_time % 10 == 0:
                    print(f"\n=== Time {current_time} Status ===")
                    completed_jobs = sum(1 for job in current_dataset.jobs 
                                       if job['completed_operations'] == len(job['operations']))
                    in_progress = sum(1 for machine in current_dataset.machines 
                                    if machine['status'] == 'busy')
                    print(f"Completed jobs: {completed_jobs}")
                    print(f"Machines busy: {in_progress}")
                
                # Run scheduling step
                action, current_dataset = run_scheduling_step(
                    current_dataset, 
                    current_time, 
                    gnn_model, 
                    heuristic_algo
                )

                # Update system
                current_dataset = update_system_state(current_dataset, action, current_time)
                
                # Check if all jobs are complete
                all_complete = all(job['completed_operations'] == len(job['operations']) 
                                 for job in current_dataset.jobs)
                if all_complete:
                    print(f"\nAll jobs completed at time {current_time}!")
                    break
                
                current_time += 1
            
            # Print final statistics
            completed_jobs = sum(1 for job in current_dataset.jobs 
                               if job['completed_operations'] == len(job['operations']))
            total_operations = sum(len(job['operations']) for job in current_dataset.jobs)
            completed_operations = sum(job['completed_operations'] for job in current_dataset.jobs)
            
            print("\nFinal Statistics:")
            print(f"Total time steps: {current_time}")
            print(f"Completed jobs: {completed_jobs}/{len(current_dataset.jobs)}")
            print(f"Completed operations: {completed_operations}/{total_operations}")
            print("Machine utilization:")
            for machine in current_dataset.machines:
                busy_time = sum(1 for job in current_dataset.jobs 
                              for op in job['operations'] 
                              if op['machine'] == machine['machine_id'] and op['status'] == 'completed')
                utilization = (busy_time / current_time) * 100 if current_time > 0 else 0
                print(f"- Machine {machine['machine_id']}: {utilization:.1f}% utilized")
    # --- Test Loop ---
    test_files = os.listdir(test_data_dir)
    for filename in test_files:
        loaded_generator = FMSDatasetGenerator.load_dataset(os.path.join(test_data_dir,filename))
        # Dummy data for FMSDatasetGenerator
        dummy_instance = FMSDatasetGenerator(num_jobs=len(loaded_generator["jobs"]),
                                            num_machines=len(loaded_generator["machines"]),
                                            num_ops_per_job_range=(3, 8), processing_time_range=(2, 20),
                                            failure_rate=0.05, setup_time_range=(0, 3), resource_capacity=6,
                                            load_factor=0.6)
        dummy_instance.jobs = loaded_generator["jobs"]
        dummy_instance.machines = loaded_generator["machines"]
        dummy_instance.resources = loaded_generator["resources"]

        current_time = 0
        num_steps = 100
        current_dataset = dummy_instance
        while current_time < num_steps:
            action, current_dataset = run_scheduling_step(current_dataset, current_time, gnn_model, heuristic_algo)
            # update the system state based on the action, compute evaluation metrics
            current_dataset = update_system_state(current_dataset, action, current_time)
            current_time += 1


class SmartHeuristic():
    def __init__(self, params=None):
        self.params = params
        self.waiting_time_weight = 0.3
        self.priority_weight = 0.4
        self.processing_time_weight = 0.3
        
    def __call__(self, gnn_embeddings, raw_features, available_jobs, graph=None):
        """
        Args:
            gnn_embeddings: Graph neural network embeddings
            raw_features: Dictionary containing job, machine, and operation features
            available_jobs: List of available job IDs
            graph: NetworkX graph object (optional)
        Returns:
            str or None: Selected job key or None if no jobs available
        """
        if not available_jobs:
            return None
            
        job_scores = {}
        for job_id in available_jobs:
            job_key = f"job_{job_id}"
            job_data = raw_features["jobs"][job_key]
            
            # 1. Priority Score (1-5 from original data)
            priority_score = job_data["priority"] / 5.0  # Normalize to 0-1
            
            # 2. Waiting Time Score
            waiting_time = job_data["time_since_arrival"]
            waiting_score = min(waiting_time / 50.0, 1.0)  # Normalize to 0-1, cap at 1
            
            # 3. Processing Time Score (prefer shorter operations)
            current_op_id = int(job_data["completed_operations"])
            op_key = f"op_{job_id}_{current_op_id}"
            if op_key in raw_features["operations"]:
                op_data = raw_features["operations"][op_key]
                proc_time = op_data["processing_time"]
                # Shorter processing times get higher scores
                processing_score = 1.0 - (proc_time / 20.0)  # Normalize to 0-1
            else:
                processing_score = 0
            
            # Calculate weighted total score
            total_score = (
                self.priority_weight * priority_score +
                self.waiting_time_weight * waiting_score +
                self.processing_time_weight * processing_score
            )
            
            job_scores[job_key] = total_score
            
            # Print decision factors for transparency
            print(f"\nScoring {job_key}:")
            print(f"  Priority Score: {priority_score:.2f}")
            print(f"  Waiting Score: {waiting_score:.2f}")
            print(f"  Processing Score: {processing_score:.2f}")
            print(f"  Total Score: {total_score:.2f}")
        
        if job_scores:
            selected_job = max(job_scores, key=job_scores.get)
            print(f"\nSelected {selected_job} with highest score!")
            return selected_job
        return None


# --- Main ---
my_heuristic = SmartHeuristic() # or use CentralityHeuristic
#my_heuristic = MyHeuristic(params = None) # Dummy Implementation
# --- Setup and Run ---
input_dim = 5  # This should match max_feature_size from prepare_gnn_input
hidden_dim = 64
output_dim = 32

# Initialize and move model to device
my_gnn_model = MyGCN(input_dim, hidden_dim, output_dim)
my_gnn_model.to(device) # Move model to GPU (or CPU if no GPU)


train_data_dir = 'fms_train_data'
val_data_dir = 'fms_val_data'
test_data_dir = 'fms_test_data'
main_simulation(train_data_dir,val_data_dir, test_data_dir, my_gnn_model, my_heuristic, train_iters = 1)
```

```python
# --- GNN Model Definition ---
class GCNLayer(nn.Module):
    def __init__(self, in_features, out_features):
        super(GCNLayer, self).__init__()
        self.linear = nn.Linear(in_features, out_features)

    def forward(self, node_features, adj_matrix):
        # Normalize adjacency matrix
        deg = torch.sum(adj_matrix, dim=1)
        deg_inv_sqrt = torch.pow(deg + 1e-6, -0.5)
        norm_adj = torch.mul(
            torch.mul(adj_matrix, deg_inv_sqrt.unsqueeze(1)),
            deg_inv_sqrt.unsqueeze(0)
        )
        
        # Perform graph convolution
        aggregated_features = torch.matmul(norm_adj, node_features)
        transformed_features = self.linear(aggregated_features)
        return transformed_features

class MyGCN(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(MyGCN, self).__init__()
        self.gcn1 = GCNLayer(input_dim, hidden_dim)
        self.gcn2 = GCNLayer(hidden_dim, output_dim)
        self.dropout = nn.Dropout(0.2)
        
    def forward(self, adj_matrix, node_features):
        # First GCN layer with ReLU and dropout
        h1 = F.relu(self.gcn1(node_features, adj_matrix))
        h1 = self.dropout(h1)
        # Second GCN layer
        h2 = self.gcn2(h1, adj_matrix)
        return h2
```

```python
# --- FMS Data Generator ---
class FMSDatasetGenerator:
    """Generates datasets for an FMS scheduling problem."""

    def __init__(self, num_jobs, num_machines, num_ops_per_job_range, processing_time_range,
                 failure_rate, setup_time_range, resource_capacity, load_factor,
                 dynamic_arrival=True, fixed_routing=False):
        self.num_jobs = num_jobs if isinstance(num_jobs, int) else random.randint(*num_jobs)
        self.num_machines = num_machines
        self.num_ops_per_job_range = num_ops_per_job_range
        self.processing_time_range = processing_time_range
        self.failure_rate = failure_rate
        self.setup_time_range = setup_time_range
        self.resource_capacity = resource_capacity
        self.load_factor = load_factor
        self.dynamic_arrival = dynamic_arrival
        self.fixed_routing = fixed_routing
        self.machines = self._generate_machines()
        self.jobs = self._generate_jobs()
        self.resources = self._generate_resources()

    def _generate_resources(self):
        resources = {
            "resource_capacity": self.resource_capacity,
            "machines":  random.sample(self.machines, int(len(self.machines) * self.load_factor))
        }
        return resources

    def _generate_machines(self):
        machines = []
        for machine_id in range(self.num_machines):
            machines.append({"machine_id": machine_id,
                               "status": "available",
                               "current_job": None,
                               "next_available_time": 0,
                               "type": random.choice(["general", "special"])
                              })
        return machines

    def _generate_jobs(self):
        jobs = []
        for job_id in range(self.num_jobs):
            num_ops = random.randint(*self.num_ops_per_job_range)
            ops = self._generate_job_ops(job_id, num_ops)
            jobs.append({"job_id": job_id, "operations": ops,
                           "arrival_time": 0 if not self.dynamic_arrival else random.randint(0,50),
                           "completed_operations":0,
                           "priority": random.randint(1,5),
                           "machine_seq": [op["machine"] for op in ops],
                           "current_operation":0
                           })
        return jobs

    def _generate_job_ops(self, job_id, num_ops):
        ops = []
        available_machines = self._get_available_machines()
        if self.fixed_routing:
            machine_seq = random.sample(available_machines, num_ops)
        else:
            machine_seq = []

        for op_id in range(num_ops):
            if self.fixed_routing:
                machine = machine_seq[op_id]
            else:
                machine = random.choice(available_machines)
            proc_time = self._triangular_distribution(*self.processing_time_range)
            setup_time = random.uniform(*self.setup_time_range) if op_id > 0 else 0.0
            ops.append({"job_id": job_id, "op_id": op_id, "machine": machine,
                            "processing_time": proc_time, "setup_time": setup_time,
                            "status": "pending"
                            })
        return ops

    def _triangular_distribution(self, min, max):
        mode = (min + max)/2
        u = random.uniform(0,1)
        if u <= (mode-min)/(max-min):
            return min + np.sqrt(u*(max-min)*(mode-min))
        else:
            return max-np.sqrt((1-u)*(max-min)*(max-mode))

    def _get_available_machines(self):
        available_machines = []
        for m in self.machines:
            available_machines.append(m["machine_id"])
        return available_machines

    def save_dataset(self, filename, save_dir):
        os.makedirs(save_dir, exist_ok = True)
        dataset = {
            "jobs": self.jobs,
            "machines": self.machines,
            "resources": self.resources,
        }

        with open(os.path.join(save_dir,filename), 'w') as f:
            json.dump(dataset, f, indent=4)

    @classmethod
    def load_dataset(cls, filename):
        with open(filename, 'r') as f:
            dataset = json.load(f)
        return dataset
```

```python
# --- Graph and Feature Preparation ---
def create_graph_and_features(dataset, current_time = 0):
    jobs = dataset.jobs
    machines = dataset.machines
    graph = nx.DiGraph()
    heuristic_data = {"jobs": {}, "machines": {}, "operations": {}}
    # 1. Node generation:
    # Add job nodes
    for job in jobs:
        graph.add_node(f"job_{job['job_id']}", type="job", **job)
        heuristic_data["jobs"][f"job_{job['job_id']}"] = _extract_job_features(job,current_time)

    # Add machine nodes
    for machine in machines:
        machine_node_data = {k: v for k, v in machine.items() if k not in ["type", "current_job"]}
        graph.add_node(f"machine_{machine['machine_id']}", type="machine", **machine_node_data)
        heuristic_data["machines"][f"machine_{machine['machine_id']}"] = _extract_machine_features(machine, current_time)

    # Add operation nodes
    for job in jobs:
        for op in job["operations"]:
            graph.add_node(f"op_{job['job_id']}_{op['op_id']}", type="operation", **op)
            heuristic_data["operations"][f"op_{job['job_id']}_{op['op_id']}"] = _extract_op_features(op, job, current_time)

    # 2. Edge Generation
    #Edges for operations in a job sequence
    for job in jobs:
        ops = job["operations"]
        for i in range(len(ops)-1):
            graph.add_edge(f"op_{job['job_id']}_{ops[i]['op_id']}",
                        f"op_{job['job_id']}_{ops[i+1]['op_id']}", type="op_seq")

    #Edges between operations and machines
    for job in jobs:
        for op in job["operations"]:
            machine_node = f"machine_{op['machine']}"
            graph.add_edge(f"op_{job['job_id']}_{op['op_id']}", machine_node, type="machine_link")

    #Disjunctive edges for operations in a machine
    for m in machines:
        ops_in_machine = []
        for job in jobs:
            for op in job["operations"]:
                if op["machine"] == m["machine_id"]:
                    ops_in_machine.append(f"op_{job['job_id']}_{op['op_id']}")
        for i in range(len(ops_in_machine)):
            for j in range(i+1, len(ops_in_machine)):
                graph.add_edge(ops_in_machine[i], ops_in_machine[j], type = "disjunctive_link")

    return graph, heuristic_data

def _extract_job_features(job, current_time):
    time_since_arrival = current_time-job["arrival_time"]
    remaining_operations = len(job["operations"])- job["completed_operations"]
    features = {"job_id": float(job["job_id"]),
                "priority": float(job["priority"]),
                "time_since_arrival": float(time_since_arrival),
                "remaining_operations":float(remaining_operations),
                "completed_operations": float(job["completed_operations"])
                }
    return features

def _extract_machine_features(machine, current_time):
    features = { "machine_id": float(machine["machine_id"]),
                "next_available_time": float(machine["next_available_time"] - current_time )
            }
    return features

def _extract_op_features(op, job, current_time):
        features = { "job_id": float(job["job_id"]),
                    "op_id": float(op["op_id"]),
                    "machine": float(op["machine"]),
                    "processing_time": float(op["processing_time"]),
                    "setup_time": float(op["setup_time"]),

        }
        return features

def prepare_gnn_input(graph, feature_data):
    # Define feature sizes for each node type
    job_feature_size = 5  # job_id, priority, time_since_arrival, remaining_ops, completed_ops
    machine_feature_size = 2  # machine_id, next_available_time
    operation_feature_size = 5  # job_id, op_id, machine, processing_time, setup_time
    
    # Use the maximum feature size for all nodes to ensure consistent dimensions
    max_feature_size = max(job_feature_size, machine_feature_size, operation_feature_size)
    
    node_features = []
    node_ids = []
    
    # Process all nodes and pad features to max_feature_size
    for node_id, node_data in graph.nodes(data=True):
        node_ids.append(node_id)
        
        if node_data["type"] == "job":
            features = list(feature_data["jobs"][f"job_{node_data['job_id']}"].values())
        elif node_data["type"] == "machine":
            features = list(feature_data["machines"][f"machine_{node_data['machine_id']}"].values())
        elif node_data["type"] == "operation":
            features = list(feature_data["operations"][f"op_{node_data['job_id']}_{node_data['op_id']}"].values())
        
        # Pad with zeros to reach max_feature_size
        padded_features = features + [0.0] * (max_feature_size - len(features))
        node_features.append(padded_features)
    
    # Convert to tensors
    node_features_tensor = torch.tensor(node_features, dtype=torch.float32)
    adj_matrix = nx.to_numpy_array(graph, nodelist=node_ids)
    adj_matrix_tensor = torch.tensor(adj_matrix, dtype=torch.float32)
    
    return node_features_tensor, adj_matrix_tensor
```

```python
# --- Loss Function and Evaluation Metrics ---
def calculate_loss(dataset, gnn_embeddings, selected_job, raw_features, current_time):
    if not selected_job:
        return torch.tensor(0.0, device=device)

    job_id = int(selected_job.split("_")[1])
    job_data = None
    for job in dataset.jobs:
        if job["job_id"] == job_id:
            job_data = job
            break

    if not job_data:
        return torch.tensor(0.0, device=device)

    current_op_id = job_data["current_operation"]
    op_key = f"op_{job_id}_{current_op_id}"
    
    # Ensure op_key is in raw_features["operations"] before accessing it
    if op_key not in raw_features["operations"]:
        return torch.tensor(0.0, device=device)

    op_data = raw_features["operations"][op_key]
    processing_time = op_data["processing_time"]

    # Calculate waiting time
    waiting_time = current_time - job_data["arrival_time"]
    for i in range(current_op_id):
        prev_op_key = f"op_{job_id}_{i}"
        if prev_op_key in raw_features["operations"]:
            waiting_time += raw_features["operations"][prev_op_key]["processing_time"]

    # Normalize waiting time and processing time
    max_waiting_time = 50.0  # Normalization factor for waiting time
    max_processing_time = 20.0  # Normalization factor for processing time

    normalized_waiting_time = min(waiting_time / max_waiting_time, 1.0)
    normalized_processing_time = processing_time / max_processing_time

    # Calculate loss based on waiting time and processing time
    loss = normalized_waiting_time + normalized_processing_time

    return torch.tensor(loss, device=device)

def calculate_makespan(dataset):
    """Calculates the makespan of the schedule."""
    max_completion_time = 0
    for job in dataset.jobs:
        job_completion_time = 0
        for op in job["operations"]:
            job_completion_time += op["processing_time"]
        max_completion_time = max(max_completion_time, job_completion_time)
    return max_completion_time

def calculate_machine_utilization(dataset, current_time):
    """Calculates the average machine utilization."""
    total_utilization = 0
    for machine in dataset.machines:
        busy_time = 0
        for job in dataset.jobs:
            for op in job["operations"]:
                if op["machine"] == machine["machine_id"] and op["status"] == "completed":
                    busy_time += op["processing_time"]

        utilization = (busy_time / current_time) if current_time > 0 else 0
        total_utilization += utilization

    avg_utilization = total_utilization / len(dataset.machines) if len(dataset.machines) > 0 else 0
    return avg_utilization
```

```python
import pandas as pd
# --- System Update and Simulation Loop ---
def update_system_state(dataset, selected_job, current_time):
    # ... (no changes in this function)
    if selected_job is None:
        for machine in dataset.machines:
            if machine['next_available_time'] <= current_time:
                machine['status'] = 'available'
                machine['current_job'] = None
        return dataset

    job_id = int(selected_job.split('_')[1])

    for job in dataset.jobs:
        if job['job_id'] == job_id:
            if job['current_operation'] >= len(job['operations']):
                return dataset

            current_op = job['operations'][job['current_operation']]
            machine_id = current_op['machine']

            for machine in dataset.machines:
                if machine['machine_id'] == machine_id and machine['status'] == 'available':
                    processing_time = current_op['processing_time']
                    setup_time = current_op['setup_time']
                    total_time = processing_time + setup_time

                    machine['status'] = 'busy'
                    machine['current_job'] = job_id
                    machine['next_available_time'] = current_time + total_time

                    current_op['status'] = 'completed'

                    job['current_operation'] += 1
                    job['completed_operations'] += 1
                    break

    for machine in dataset.machines:
        if machine['next_available_time'] <= current_time:
            machine['status'] = 'available'
            machine['current_job'] = None

    return dataset

def run_scheduling_step(dataset, current_time, gnn_model, heuristic_algo, optimizer, loss_history, makespan_history, utilization_history, update_interval,batch_idx, metrics):
    graph, feature_data = create_graph_and_features(dataset, current_time)
    available_jobs = [job["job_id"] for job in dataset.jobs if job["current_operation"] < len(job["operations"])]

    node_features, adj_matrix = prepare_gnn_input(graph, feature_data)

    node_features = node_features.to(device)
    adj_matrix = adj_matrix.to(device)

    gnn_embeddings = gnn_model(adj_matrix, node_features)

    action = heuristic_algo(gnn_embeddings, feature_data, available_jobs, graph)

    # Calculate loss and update GNN model
    loss = calculate_loss(dataset, gnn_embeddings, action, feature_data, current_time)

    if loss is not None and loss.requires_grad:
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    # Visualization update (every 10 batches)
    if batch_idx is not None and batch_idx % 10 == 0:
        makespan = calculate_makespan(dataset)
        utilization = calculate_machine_utilization(dataset, current_time)
        loss_history.append(loss.item())
        makespan_history.append(makespan)
        utilization_history.append(utilization)
        plot_comparison_metrics(metrics, 10) # Update the plot every 10 batches

    # Clear tensors that are not needed anymore
    del node_features
    del adj_matrix

    return action, dataset

def main_simulation(train_data_dir, val_data_dir, test_data_dir, gnn_model, heuristic_algos, config):
    batch_size = config["batch_size"]
    optimizers = {name: optim.Adam(gnn_model.parameters(), lr=config["learning_rate"], weight_decay=config["weight_decay"]) for name in heuristic_algos}
    metrics = {name: {"loss_history": [], "makespan_history": [], "utilization_history": []} for name in heuristic_algos}

    for i in range(config["train_iters"]):
        for heuristic_name, heuristic_algo in heuristic_algos.items():
            print(f"Running with {heuristic_name} heuristic")
            dataset_files = [os.path.join(train_data_dir, f) for f in os.listdir(train_data_dir) if os.path.isfile(os.path.join(train_data_dir, f))]
            num_batches = (len(dataset_files) + batch_size - 1) // batch_size

            for batch_idx in range(num_batches):
                print(f"Processing batch {batch_idx + 1}/{num_batches} of datasets")
                batch_files = dataset_files[batch_idx * batch_size : (batch_idx + 1) * batch_size]
                batch_datasets = [FMSDatasetGenerator.load_dataset(f) for f in batch_files]

                for dataset in batch_datasets:
                    current_dataset = FMSDatasetGenerator(
                        num_jobs=len(dataset["jobs"]),
                        num_machines=len(dataset["machines"]),
                        num_ops_per_job_range=(3, 8),
                        processing_time_range=(2, 20),
                        failure_rate=0.05,
                        setup_time_range=(0, 3),
                        resource_capacity=6,
                        load_factor=0.6
                    )
                    current_dataset.jobs = dataset["jobs"]
                    current_dataset.machines = dataset["machines"]
                    current_dataset.resources = dataset["resources"]

                    current_time = 0
                    while current_time < config["num_steps"]:
                        action, current_dataset = run_scheduling_step(
                            current_dataset, current_time, gnn_model, heuristic_algo, optimizers[heuristic_name],
                            metrics[heuristic_name]["loss_history"], metrics[heuristic_name]["makespan_history"],
                            metrics[heuristic_name]["utilization_history"], config["plot_update_interval"], batch_idx , metrics
                        )
                        current_dataset = update_system_state(current_dataset, action, current_time)
                        all_complete = all(job['completed_operations'] == len(job['operations']) for job in current_dataset.jobs)
                        if all_complete:
                            print(f"\nAll jobs completed at time {current_time}!")
                            break
                        current_time += 1

                    del current_dataset  # Remove dataset after processing

                # Reset the GNN model's parameters for the next batch
                my_gnn_model = MyGCN(config["input_dim"], config["hidden_dim"], config["output_dim"]).to(device)
                optimizers[heuristic_name] = optim.Adam(my_gnn_model.parameters(), lr=config["learning_rate"], weight_decay=config["weight_decay"])

            # Reset the GNN model's parameters for the next heuristic
            my_gnn_model = MyGCN(config["input_dim"], config["hidden_dim"], config["output_dim"]).to(device)
            optimizers[heuristic_name] = optim.Adam(my_gnn_model.parameters(), lr=config["learning_rate"], weight_decay=config["weight_decay"])

    # --- Test Loop ---
    test_files = os.listdir(test_data_dir)
    for filename in test_files:
        loaded_generator = FMSDatasetGenerator.load_dataset(os.path.join(test_data_dir, filename))

        for heuristic_name, heuristic_algo in heuristic_algos.items():
            print(f"Testing with {heuristic_name} heuristic")
            current_dataset = FMSDatasetGenerator(
                num_jobs=len(loaded_generator["jobs"]),
                num_machines=len(loaded_generator["machines"]),
                num_ops_per_job_range=(3, 8),
                processing_time_range=(2, 20),
                failure_rate=0.05,
                setup_time_range=(0, 3),
                resource_capacity=6,
                load_factor=0.6
            )
            current_dataset.jobs = loaded_generator["jobs"]
            current_dataset.machines = loaded_generator["machines"]
            current_dataset.resources = loaded_generator["resources"]

            current_time = 0
            while current_time < config["num_steps"]:
                action, current_dataset = run_scheduling_step(
                    current_dataset, current_time, gnn_model, heuristic_algo, optimizers[heuristic_name],
                    metrics[heuristic_name]["loss_history"], metrics[heuristic_name]["makespan_history"],
                    metrics[heuristic_name]["utilization_history"], config["plot_update_interval"]
                )
                current_dataset = update_system_state(current_dataset, action, current_time)
                current_time += 1

            del current_dataset
            # Reset the GNN model's parameters for the next heuristic
            my_gnn_model = MyGCN(config["input_dim"], config["hidden_dim"], config["output_dim"]).to(device)
            optimizers[heuristic_name] = optim.Adam(my_gnn_model.parameters(), lr=config["learning_rate"], weight_decay=config["weight_decay"])

    # --- Visualization ---
    plot_comparison_metrics(metrics, config["plot_update_interval"])

    # Visualize final embeddings for each heuristic
    for heuristic_name, heuristic_algo in heuristic_algos.items():
        graph, feature_data = create_graph_and_features(current_dataset, current_time)
        node_features, adj_matrix = prepare_gnn_input(graph, feature_data)
        node_features = node_features.to(device)
        adj_matrix = adj_matrix.to(device)
        with torch.no_grad():
            final_embeddings = gnn_model(adj_matrix, node_features).cpu().numpy()
        print(f"Embedding visualization for {heuristic_name} heuristic")
        visualize_embeddings(final_embeddings, graph)

def plot_comparison_metrics(metrics, update_interval):
    """Plots the comparison of loss, makespan distribution, and utilization distribution for different heuristics."""
    clear_output(wait=True)

    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    for heuristic_name, metric in metrics.items():
        # Plot Loss (as a line graph)
        axes[0].plot(metric["loss_history"], label=f"{heuristic_name} Loss")
        axes[0].set_xlabel(f"Update Step (x{update_interval})")
        axes[0].set_ylabel("Loss")
        axes[0].set_title("Training Loss Comparison")
        axes[0].legend()

        # Plot Makespan Distribution (as a histogram)
        axes[1].hist(metric["makespan_history"], label=f"{heuristic_name} Makespan", bins=20, alpha=0.7)
        axes[1].set_xlabel("Makespan")
        axes[1].set_ylabel("Frequency")
        axes[1].set_title("Makespan Distribution")
        axes[1].legend()

        # Plot Utilization Distribution (as a histogram)
        axes[2].hist(metric["utilization_history"], label=f"{heuristic_name} Utilization", bins=20, alpha=0.7)
        axes[2].set_xlabel("Average Utilization")
        axes[2].set_ylabel("Frequency")
        axes[2].set_title("Machine Utilization Distribution")
        axes[2].legend()

    plt.tight_layout()
    plt.show()
```
